### Ki Kare Testi

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
df.dropna(inplace=True)
value=df.drop(['Price_in_thousands', 'Manufacturer', 'Model','Vehicle_type', 'Latest_Launch'],axis=1)
target=df['Price_in_thousands']

value = value.astype(int)
target=target.astype(int)
value
```

```python
chi2_features = SelectKBest(chi2, k = 5)
X_kbest_features = chi2_features.fit(value, target)
features = value.columns[X_kbest_features.get_support()]
print(features)
```
