```python
def change_name(cols):
    x = cols[0]
    y = cols[1]
    if x!='Neon':
        return x
    else:
        if y == "Dodge":
            return 'Neon_d'
        else:
            return 'Neon_p'
```

```python
df['Model'] = df[['Model','Manufacturer']].apply(change_name,axis=1)
df[(df['Model'] =='Neon_d') | (df['Model'] == 'Neon_p')]
```

## Tek bir niteliğin eksik verilerini ortalama ile tamamla

```python
df["__year_resale_value"].fillna(df["__year_resale_value"].mean())  # inplace=True
```

## Nümerik niteliklerin eksik verilerini ortalama değerleri ile tamamla

```python
df.fillna(df.mean(numeric_only=True))
```

## Bir niteliği başka bir niteliğe göre gruplandırıp ortalama ile tamamlama

```python
df["__year_resale_value"]=df["__year_resale_value"].fillna(df.groupby("Manufacturer")["__year_resale_value"].mean())
```

## Kategorik bir değişkenin eksik verilerini tamamlama

```python
df["Manufacturer"].fillna(df["Manufacturer"].mode()[0])
```

```python
Q1 = df_table.quantile(0.25)
Q2 = df_table.quantile(0.5)
Q3 = df_table.quantile(0.75)
IQR = Q3 - Q1    
# Çeyreklik Açıklığı 
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR 
print("Alt Sınır Değeri : " + str(lower_bound)) 
print("Üst Sınır Değeri : " + str(upper_bound)) 
print("1. Çeyrek : " + str(Q1)) 
print("2. Çeyrek : " + str(Q2))
print("3. Çeyrek : "+ str(Q3)) 
```

```python
# Bu değerler bir diziye atılıyor 
outliers_vector = (df_table < (lower_bound)) | (df_table > (upper_bound) ) 
```

```python
clean_df_table = df_table[~((df_table<(lower_bound)) | (df_table > (upper_bound)))] 
clean_df_table.shape
```

### Ki Kare Testi

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
df.dropna(inplace=True)
value=df.drop(['Price_in_thousands', 'Manufacturer', 'Model','Vehicle_type', 'Latest_Launch'],axis=1)
target=df['Price_in_thousands']

value = value.astype(int)
target=target.astype(int)
value
```

```python
chi2_features = SelectKBest(chi2, k = 5)
X_kbest_features = chi2_features.fit(value, target)
features = value.columns[X_kbest_features.get_support()]
print(features)
```
