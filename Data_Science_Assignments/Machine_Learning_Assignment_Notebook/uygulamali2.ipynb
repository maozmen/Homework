## 3) Ön İşleme Yapılması

```python
data = data.loc[:, ~data.columns.isin(['ID', 'year', 'rate_of_interest', 'Interest_rate_spread', 'Upfront_charges'])]
data.shape[1]
```

ID, year, rate_of_interest, interest_rate_spread, upfront_charges niteliklerini veri setinden at. ID Sınıflandırmada önemsiz. Year yalnızca 2019 değerine sahip. Biz sınıflandırmada kredi verilip verilmediğini inceleyeceğiz ama rate_of_interest, interest_rate_spread ve upfront_charges nitelikleri kredisi onaylananların kredi özelliklerini tanımlıyor.

```python
data.loc[data['LTV'].isna(), 'LTV'] = data['LTV'].mean()
data
```

LTV Niteliği null olan kayıtların LTV niteliğine niteliğin tüm kayıtlardaki ortalama değerini ata.

```python
data.loc[data['dtir1'].isna(), 'dtir1'] = data['dtir1'].mean()
```

dtir1 Niteliği null olan kayıtların dtir1 niteliğine niteliğin tüm kayıtlardaki ortalama değerini ata.

```python
series_pro_na = data['property_value'].isna()

data.loc[series_pro_na, 'property_value'] = data[series_pro_na]['loan_amount'] / data[series_pro_na]['LTV']

data['property_value'].isna().sum()
data['loan_amount'].isna().sum()
```

Property value niteliği null kayıtların property value niteliğine loan amount niteliği / LTV niteliği değerini ata.

```python
data = data.dropna()
data.shape[0]
```

Hala null değeri olan kayıtları at. 148670 Kayıttan 135108'sı kalmış.

```python
data_no_out = data.copy()

def drop_outliers(src, dst, column):
    """
    returns modified destination data
    """
    q1 = src[column].quantile(0.25)
    q3 = src[column].quantile(0.75)
    iqr = q3 - q1
    low = q1 - 1.5*iqr
    high = q3 + 1.5*iqr
    return dst[(dst[column] <= high) & (dst[column] >= low)]

for clmn in ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']:
    data_no_out = drop_outliers(data, data_no_out, clmn)
    print(f"{clmn:14} niteliğindeki aykırı değerler atıldıktan sonra kalan kayıt sayısı: {data_no_out.shape[0]}")

data_no_out.shape[0]
```

Veri setindeki aykırı değere sahip kayıtları at. 135108 Kayıttan 118496'ü kalmış.

## 5) Veri Setini Eğitim Ve Test Seti Olarak Ayır

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_no_out.loc[:, ~data_no_out.columns.isin(['Status'])],\
                                                    data_no_out.loc[:, 'Status'], train_size = 0.7, random_state = 0)
print(f"Eğitim setindeki kayıt sayısı: {x_train.shape[0]}\nTest setindeki kayıt sayısı: {x_test.shape[0]}")
```

## 6) Makine Öğrenmesi Sınıflandırma Algoritmaları İle Modellerin Eğitilmesi Ve Kaydedilmesi

## 6.1) Naive Bayes Modelinin Eğitilmesi Ve Kaydedilmesi

```python
x_train_nb = x_train.copy()
x_test_nb = x_test.copy()

x_train_nb = x_train_nb.loc[:, ~x_train_nb.columns.isin(['property_value'])]
x_test_nb = x_test_nb.loc[:, ~x_test_nb.columns.isin(['property_value'])]
```

Property value niteliği, LTV ve loan amount niteliklerinden elde edilebildiği için naive bayes'te kullanılacak niteliklerden property value'yu çıkar.

```python
for i, clmn in enumerate(['loan_amount', 'income', 'Credit_Score', 'LTV', 'dtir1']):
    _, bin_ = pd.qcut(x_train_nb.loc[:,clmn], 10, retbins = True)
    bin_[0] = min(x_train_nb[clmn].min(), x_test_nb[clmn].min()) - 1 #categorilerin en alt sınırını gerekiyorsa genişlet
    bin_[-1] = max(x_train_nb[clmn].max(), x_test_nb[clmn].max()) + 1 #categorilerin en üst sınırını gerekiyorsa genişlet
    x_train_nb.loc[:, clmn] = pd.cut(x_train_nb.loc[:,clmn], bin_)
    x_test_nb.loc[:, clmn] = pd.cut(x_test_nb.loc[:,clmn], bin_)
    
x_train_nb.loc[x_train_nb['term'] != 360.0, 'term'] = '!360'
x_train_nb.loc[x_train_nb['term'] == 360.0, 'term'] = '360'

x_test_nb.loc[x_test_nb['term'] != 360.0, 'term'] = '!360'
x_test_nb.loc[x_test_nb['term'] == 360.0, 'term'] = '360'
```

loan amount, income, credit score, LTV, dtir1 Niteliklerinin her birini 10 kategoriye böl. term Niteliğini kaydın değeri 360'a eşit olup olmamasına göre ikiye böl.

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
x_train_nb = enc.fit_transform(x_train_nb)
x_test_nb = enc.transform(x_test_nb)
```

Veri setinde hepsi kategorik olan nitelikleri sklearn kütüphanesindeki CategoricalNB algoritmasının istediği hale getirebilmek için sayısal olarak ifade et.

```python
from sklearn.naive_bayes import CategoricalNB

cnb = CategoricalNB()
cnb.fit(x_train_nb, y_train)
```

Naive bayes modelini eğit.

```python
import pickle

with open("naive_bayes.pkl", 'wb') as f:
    pickle.dump(cnb, f)
```

Naive bayes modelini kaydet.

## 6.2) Karar Ağacı Modelinin Eğitilmesi Ve Kaydedilmesi

```python
x_train_ka = x_train.copy()
x_test_ka = x_test.copy()

for i, clmn in enumerate(['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']):
    _, bin_ = pd.qcut(x_train_ka.loc[:,clmn], 10, retbins = True)
    bin_[0] = min(x_train_ka[clmn].min(), x_test_ka[clmn].min()) - 1 #categorilerin en alt sınırını gerekiyorsa genişlet
    bin_[-1] = max(x_train_ka[clmn].max(), x_test_ka[clmn].max()) + 1 #categorilerin en üst sınırını gerekiyorsa genişlet
    x_train_ka.loc[:, clmn] = pd.cut(x_train_ka.loc[:,clmn], bin_)
    x_test_ka.loc[:, clmn] = pd.cut(x_test_ka.loc[:,clmn], bin_)
    
x_train_ka.loc[x_train_ka['term'] != 360.0, 'term'] = '!360'
x_train_ka.loc[x_train_ka['term'] == 360.0, 'term'] = '360'

x_test_ka.loc[x_test_ka['term'] != 360.0, 'term'] = '!360'
x_test_ka.loc[x_test_ka['term'] == 360.0, 'term'] = '360'
```

loan amount, property value, income, credit score, LTV, dtir1 Niteliklerinin her birini 10 kategoriye böl. term Niteliğini kaydın değeri 360'a eşit olup olmamasına göre ikiye böl.

```python
enc = OrdinalEncoder()
x_train_ka = enc.fit_transform(x_train_ka)
x_test_ka = enc.transform(x_test_ka)
```

Veri setinde hepsi kategorik olan nitelikleri sklearn kütüphanesindeki DecisionTreeClassifier algoritmasının istediği hale getirebilmek için sayısal olarak ifade et.

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth = x_train_ka.shape[1] // 2)
dtc.fit(x_train_ka, y_train)
```

Karar ağacı modelini maksimum derinlik nitelik sayısının yarısı olacak şekilde oluştur ve modeli eğit.

```python
with open("karar_agaci.pkl", 'wb') as f:
    pickle.dump(dtc, f)
```

Karar ağacı modelini kaydet.

## 6.3) KNN Modelinin Eğitilmesi Ve Kaydedilmesi

```python
data_knn = data_no_out.copy()
```

KNN Modelinde one hot encoding yapmak uygun olacağından ve one hot encodingi eğitim ve test seti ayrılmadan yapmak daha elverişli olduğundan veri ön işleme yaptıktan sonraki dataframe ile devam ediyoruz.

```python
data_knn.loc[data_knn['term'] != 360.0, 'term'] = '!360'
data_knn.loc[data_knn['term'] == 360.0, 'term'] = '360'
```

term Niteliğini kaydın değeri 360'a eşit olup olmamasına göre ikiye böl.

```python
dummies = pd.get_dummies(data_knn.loc[:, ~data_knn.columns.isin(['loan_amount', 'property_value',\
                                                                 'income', 'Credit_Score', 'LTV', 'dtir1', 'Status'])])

data_knn = data_knn.loc[:, data_knn.columns.isin(['loan_amount', 'property_value', 'income',\
                                                  'Credit_Score', 'LTV', 'dtir1', 'Status'])]

data_knn = pd.concat([data_knn, dummies], axis = 1)
```

Kategorik nitelikleri one hot encoding olarak ifade et.

```python
x_train_knn, x_test_knn, y_train, y_test = train_test_split(data_knn.loc[:, ~data_knn.columns.isin(['Status'])],\
                                                            data_knn.loc[:, 'Status'], train_size = 0.7, random_state = 0)
```

Veri setini eğitim seti %70, test seti %30 olacak şekilde ayır.

```python
min_ = x_train_knn.loc[:, ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']].min()
max_ = x_train_knn.loc[:, ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']].max()

x_train_con = x_train.loc[:, ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']]
x_train_con = (x_train_con - min_)/(max_ - min_)
x_train_knn.loc[:, ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']] = x_train_con

x_test_con = x_test.loc[:, ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']]
x_test_con = (x_test_con - min_)/(max_ - min_)
x_test_knn.loc[:, ['loan_amount', 'property_value', 'income', 'Credit_Score', 'LTV', 'dtir1']] = x_test_con
```

Eğitim setindeki sürekli niteliklerin değerlerini [0,1] aralığına çek. Test setini de eğitim setindeki değerlere göre normalize et.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train_knn, y_train)
```

KNN Modelini komşu sayısı 5 olacak şekilde oluştur ve modeli eğit.

```python
with open("knn.pkl", 'wb') as f:
    pickle.dump(knn, f)
```

KNN Modelini kaydet.
